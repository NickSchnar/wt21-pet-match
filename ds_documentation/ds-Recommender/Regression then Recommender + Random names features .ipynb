```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
features = ['picked_up', 'adapts_quickly', 'likes_stroke', 'friendly']
```

## note: this Dataset only works on 25% (i need to find a way to make work on the whole one)

```python
### for convinance i took the 25% of the sample test for pridection 

dfRecommend = pd.concat([compare, df[['picked_up','adapts_quickly','likes_stroke','friendly']]], axis = 1).reindex(compare.index)

#dfRecommend.isna().value_counts()

dfRecommend["cat_index"]= dfRecommend.index
dfRecommend
```

```python

def combine_features(row):
    return str(row['Predicted_cat'])+" "+str(row['picked_up'])+" "+str(row['adapts_quickly'])+" "+str(row['likes_stroke'])+" "+str(row['friendly'])
```

```python
for feature in features:
    dfRecommend[feature] = dfRecommend[feature].fillna('') #filling all NaNs with blank string
    


dfRecommend["combined_features"] = dfRecommend.apply(combine_features,axis=1) #applying combined_features() method over each rows of dataframe and storing the combined string in "combined_features" column
dfRecommend

```

```python
cv = CountVectorizer() #creating new CountVectorizer() object
count_matrix = cv.fit_transform(dfRecommend["combined_features"]) 
```

```python
#  the cosine similarity matrix from the count matrix.
cosine_sim = cosine_similarity(count_matrix)
```

```python
def get_satisfaction_from_catIndex(cat_index):
    return dfRecommend[dfRecommend.cat_index == cat_index]["Predicted-adopt"].values[0]


def get_catIndex_from_satisfaction(Predicted_adopt):
    return dfRecommend[dfRecommend.Predicted_adopt == Predicted_adopt]["cat_index"].values[0]
```
